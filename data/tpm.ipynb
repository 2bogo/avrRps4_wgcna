```python
import glob
import pandas as pd
from functools import reduce
```

```python
file_list = glob.glob('./tpm/*.tsv')
```

```python
df_list = [pd.read_csv(x,sep='\t')[['target_id','est_counts']] for x in file_list]
```

```python
df_mat = reduce(lambda a,b : pd.merge(a,b, on='target_id', how = 'outer'), df_list)
```

```python
df_mat
```

```python
file_list[0]
```

```python
c = ['target_id'] + [x.split('/')[-1].split('_')[0] for x in file_list]
```

```python
df_mat.columns = c
```

```python
df_mat['target_id']
```

```python
df_mat['target_id'] = list(map(lambda x : x.split('.')[0], df_mat['target_id']))
```

```python
gene_list = sorted(set(df_mat['target_id']))
```

```python
df = pd.DataFrame()

for ix in gene_list:
    m = df_mat['target_id'] == ix
    df[ix] = df_mat[m][df_mat.columns[1:]].sum(numeric_only=True, axis=0)
```

```python
m = df_mat['target_id'] == gene_list[1]
```

```python
df = pd.DataFrame()
df[gene_list[1]] = df_mat[m][df_mat.columns[1:]].sum(numeric_only=True, axis=0)
```

```python
df = df.T
```

```python
df.to_csv('avrRps4_geneCount.csv')
```
