```python
import pandas as pd
```

```python
df = pd.read_csv('./data/avrRps4_counts.csv')
```

```python
df_ix = df.set_index('target_id')
```

```python
avr = '''SRR11510194
SRR11510195
SRR11510196
SRR7429855
SRR7429859
SRR7429873
SRR7429877
SRR7429885
SRR7429889
SRR5123620
SRR5123621
SRR5123622
SRR5123623
SRR6003795
SRR6003824
SRR6003802
SRR6003827
SRR11934411
SRR11934414
SRR11934417
SRR11934420
SRR11934425
SRR11934428
SRR11934439
SRR11934443
SRR11934445
SRR11934450
SRR11934453
SRR11934456
SRR11934459
SRR11934464
SRR11934467
SRR11934478
SRR11934482
SRR11934484'''
```

```python
avr_list = avr.split('\n')
```

```python
c = []
for x in df_ix.columns:
    if x in avr_list:
        c.append('avrRps4' + '_' + x)
    else:
        c.append('control' + '_' + x)
df_ix.columns = c

c = sorted(c)[::-1]
df_sort = df_ix[c]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_pick = df_sort.loc[['AT5G45250.1', 'AT5G45260.1']]
df_pick = df_pick.rename(index={'AT5G45250.1' : 'RPS4', 'AT5G45260.1' : 'RRS1'})
```

```python
fig, ax = plt.subplots(figsize=(40,4))
sns.heatmap(df_pick)
plt.show()
```
